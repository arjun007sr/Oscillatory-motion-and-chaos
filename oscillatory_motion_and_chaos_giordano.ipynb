**SHM (Euler's Method)**

```python
dt = 0.04
time = 10
n = int(time/dt)
w = np.zeros(n+1,float)
theta = np.zeros(n+1,float)
theta[0] = .02
t = np.zeros(n+1,float)
for i in range(n):
    w[i+1] = w[i] - (g/l)*theta[i]*dt
    theta[i+1] = theta[i] + w[i]*dt
    t[i+1] = t[i] + dt
plt.plot(t,theta)
plt.xlim(0,10)
plt.xlabel("time")
plt.ylabel("theta")
```

**Pendulum for Damped, Overdamped and Critically damped cases** FIGURE 3.4

```python
dt = 0.04
time = 10
q = 5
n = int(time/dt)
w = np.zeros(n+1,float)
theta = np.zeros(n+1,float)
theta[0] = .2
t = np.zeros(n+1,float)
def damped(q):
    for i in range(n):
        w[i+1] = w[i] - (g/l)*theta[i]*dt - q*w[i]*dt
        theta[i+1] = theta[i] + w[i+1]*dt
        t[i+1] = t[i] + dt
    return theta
plt.plot(t,damped(1))
plt.plot(t,damped(10))
plt.plot(t,damped(5))
plt.xlim(0,10)
plt.xlabel("time")
plt.ylabel("theta")
plt.legend(["underdamped","overdamped","critically damped"])
```

**Driven Pendulum (Euler Cromer)** FIGURE 3.5(a)

```python
dt = 0.04
time = 20
q = 1
omega = 2.0
FD = 0.2
n = int(time/dt)
w = np.zeros(n+1,float)
theta = np.zeros(n+1,float)
theta[0] = 0.2
t = np.zeros(n+1,float)
for i in range(n):
    w[i+1] = w[i] - (g/l)*theta[i]*dt - q*w[i]*dt + FD*np.sin(omega*t[i])*dt
    theta[i+1] = theta[i] + w[i+1]*dt
    t[i+1] = t[i] + dt
plt.title("Driven Pendulum")
plt.plot(t,theta)
plt.xlim(0,20)
plt.xlabel("time")
plt.ylabel("theta")
```

In constructing the Poincare section in Figure 3.9 we plotted points only at times that were in phase that were in phase with the drive force; that is, at times *t = 2$\pi$n / $\Omega_D$*, where n is an integer. However, we could just as easily have chosen to make the plot at times corresponding to a maximum of the drive force, or at times $\pi / 4$ out of phase with this force, etc. Construct the the Poincare sections for these cases and compare them with Figure 3.9<br>Book    : Computational Physics (2nd Edition)<br>Authors : Nicholas J. Giordano, Hisao Nakanishi<br>Chapter3: Oscillatory motion and chaos<br>Question: 3.12

```python
time = 20000
dt = 0.04
n = int(time/dt)
q = 0.5
FD = 1.2
l = 9.8
omega = 2/3
w = np.zeros(n+1,float)
theta = np.zeros(n+1,float)
theta[0] = 0.2
t = np.zeros(n+1,float)
for i in range(n):
    w[i+1] = w[i] + (-(g*np.sin(theta[i]))/l - q*w[i] + FD*np.sin(omega*t[i]))*dt
    theta[i+1] = theta[i] + w[i+1]*dt
    if (theta[i+1] < -np.pi):
        theta[i+1] = theta[i+1] + 2*np.pi
    elif  (theta[i+1] > +np.pi):
        theta[i+1] = theta[i+1] - 2*np.pi
    t[i+1] = t[i] + dt
p = 0
wnew = np.zeros(len(w),float)
thetanew = np.zeros(len(theta),float)
wnew2 = np.zeros(len(w),float)
thetanew2 = np.zeros(len(theta),float)
for j in range(len(t)):
    if ((abs(t[j] - (p*np.pi/2)/omega)) <= dt/2):
        wnew[p-1] = w[j]
        thetanew[p-1] = theta[j]
        p = p+1
f = 0
for j in range(len(t)):
    if ((abs(t[j] - (f*np.pi)/omega)) <= dt/2):
        wnew2[f-1] = w[j]
        thetanew2[f-1] = theta[j]
        f = f+1
plt.figure(figsize = (10,4))
plt.subplot(1,2,1)
plt.plot(thetanew[:p-1],wnew[:p-1],"r.")
plt.xlabel("Angular position($\Theta$)")
plt.ylabel("Angular Velocity($\omega$)")
plt.xlim(-4,4)
plt.ylim(-3,3)
plt.legend(["$\pi / 4$"])
plt.subplot(1,2,2)
plt.plot(thetanew2[:f-1],wnew2[:f-1],"g.")
plt.xlabel("Angular position($\Theta$)")
plt.ylabel("Angular Velocity($\omega$)")
plt.xlim(-4,4)
plt.ylim(-3,3)
plt.legend(["$\pi / 2$"])
plt.savefig('plot4.png', dpi=300, bbox_inches='tight', facecolor = "w")
```
